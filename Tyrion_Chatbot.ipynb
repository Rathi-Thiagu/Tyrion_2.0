# **Tyrion** - Conversational AI for companies powered by RASA 1.10.3 framework

The chatbot's primary goal is to make a company's resources more available to employees with fewer searches/conversations.

> I got motivated with this usecase based on my work experience at Infosys. It was a 377-acre complex with almost 14,000 IT workers, all working together.
Even though there are lots of resouces available, sometimes you may not have the chance to explore them, especially if you are working at clients place. Until my last day at Infosys, I had no clue there was an arts/sports club.


So, in essence, Tyrion may be viewed as an employee who has been with Infosys for more than 20 years and is well-versed in the company's workflow, facilities, and products.Tyrion will be able to assist you with any company-related questions.


# Challenges Faced:

1. Cannot use RASA 2.0 as it is not compatible with latest tensorflow version 2.x.Even when the tensorflow version is selected to 1.x , it gave an error while training "__iter__() is only supported inside of tf.function or when eager execution is enabled." which is a problem with incompatible tensorflow versions. so used RASA version 1.10.3.

2. Repeated colab session crashing and runtime disconnection due to version updates of ipython, ipykernel, tornado, prompt-toolkit or pyzmq. Resolved by executing the below command  "pip install -U ipython" and restarting the runtime.

# Observations:
   1. RASA 2.0 is more comfortable to workwith as the intensions and actions can be clearly defined in the stories.md. In addition, all of the training files in RASA 2.0 are.yml files, which may be modified just by opening the file, unlike RASA1.10, which requires command execution.
   2. RASA 2.0 project can only be downloaded after the initial project training in Rasa playground.
